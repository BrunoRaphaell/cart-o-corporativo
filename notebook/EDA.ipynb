```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
df_cartao = pd.read_csv('..\dados\cartao_corporativo_2003-2022.csv', 
                        encoding='latin-1', sep=';')
df_cartao.head()
```

```python
df_cartao.tail()
```

```python
# Removendo as duas últimas linhas do DataFrame

df_cartao.drop(labels=[df_cartao.shape[0]-1, df_cartao.shape[0]-2], axis=0, inplace=True)
```

```python
print(f"Quantidade de linhas: {df_cartao.shape[0]}")
print(f"Quantidade de colunas: {df_cartao.shape[1]}")
```

```python
print(f"Colunas: {df_cartao.columns}")
```

```python
df_cartao.dtypes
```

```python
# Convertentendo a coluna DATA PGTO para datetime

df_cartao['DATA PGTO'] = pd.to_datetime(df_cartao['DATA PGTO'], dayfirst=True)
df_cartao.dtypes
```

```python
df_cartao.head()
```

```python
df_cartao['ano'] = df_cartao['DATA PGTO'].dt.year
presidente_map = {2003: 'Lula', 2004: 'Lula', 2005: 'Lula', 2006: 'Lula', 2007: 'Lula', 2008: 'Lula', 2009: 'Lula', 2010: 'Lula', 2011: 'Lula', 2012: 'Lula', 2013: 'Dilma', 2014: 'Dilma', 2015: 'Dilma', 2016: 'Dilma', 2017: 'Temer', 2018: 'Temer', 2019: 'Bolsonaro', 2020: 'Bolsonaro', 2021: 'Bolsonaro', 2022: 'Bolsonaro'}
df_cartao['presidente'] = df_cartao['ano'].map(presidente_map)
```

```python
df_cartao[df_cartao['presidente'] == 'Bolsonaro']['VALOR']
```

```python
df_cartao[df_cartao['VALOR'].str.contains('-')]['VALOR']
```

```python
df_cartao = df_cartao[~df_cartao['VALOR'].str.contains('-')]
```

```python
texto_gasto = f"R$ {df_cartao[df_cartao['presidente'] == 'Bolsonaro']['VALOR'].sum():_.2f}"
texto_gasto = texto_gasto.replace('.', '.').replace('_', '.')
print(f"Gastos totais de bolsonaro: {texto_gasto}")
```

```python
'R$ 1.696,90'.replace('.', '').replace(',', '.')
```

```python
# Converter a coluna VALOR para float

df_cartao['VALOR'] = df_cartao['VALOR']\
    .str.replace('R\$', '', regex=True)\
    .str.replace('.', '', regex=True)\
    .str.replace(',', '.', regex=True).astype(float)
```

```python
df_cartao[df_cartao['presidente'] == 'Bolsonaro']['VALOR'].sum()
```
